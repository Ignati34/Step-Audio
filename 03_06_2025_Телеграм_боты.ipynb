## **2. Синхронность и асинхронность**

### 📊 Сравнение синхронных и асинхронных Telegram-ботов

Синхронные боты обрабатывают запросы последовательно, что может приводить к задержкам при выполнении операций ввода-вывода. Асинхронные боты позволяют обрабатывать несколько запросов одновременно, повышая эффективность и производительность.

| **Критерий**               | **Синхронный бот**                                                                 | **Асинхронный бот**                                                                 |
|:---------------------------:|:----------------------------------------------------------------------------------:|:-----------------------------------------------------------------------------------:|
| **Блокировка исполнения**   | Блокирует выполнение до завершения текущей задачи                                  | Неблокирующее выполнение, позволяет обрабатывать другие задачи параллельно          |
| **Использование потоков**   | Требует дополнительных потоков или процессов для параллельной обработки            | Использует единый поток с планировщиком задач (`asyncio`)                            |
| **Обработка запросов**      | Обрабатывает запросы последовательно                                               | Может обрабатывать несколько запросов одновременно                                  |
| **Простота реализации**     | Проще в реализации, не требует знаний асинхронного программирования                | Требует понимания `async`/`await` и работы с `asyncio`                               |
| **Производительность**      | Может испытывать задержки при высокой нагрузке или операциях ввода-вывода          | Обеспечивает лучшую производительность при работе с I/O и высокой нагрузке           |
| **Поддержка библиотек**     | Поддерживается в `pyTelegramBotAPI`, `python-telegram-bot` (до версии 13)          | Поддерживается в `python-telegram-bot` (с версии 20), `aiogram`                      |
| **Совместимость с FastAPI** | Требует дополнительных усилий для интеграции с асинхронными фреймворками          | Легко интегрируется с асинхронными фреймворками, такими как FastAPI                 |
| **Подходящие случаи**       | Подходит для простых ботов с низкой нагрузкой или когда важна простота реализации  | Идеален для ботов с высокой нагрузкой и сложной логикой                              |


## **3. Обзор популярных библиотек для создания Telegram-ботов**

### 🤖 Популярные библиотеки для создания Telegram-ботов на Python (2025)

В 2025 году для разработки Telegram-ботов на Python наиболее популярны следующие библиотеки:

---

#### 1. `python-telegram-bot`

**Описание:**  
`python-telegram-bot` — одна из самых популярных и активно развиваемых библиотек для создания Telegram-ботов на Python. С версии 20 она полностью асинхронная и совместима с `asyncio`, что делает её отличным выбором для современных приложений.

**Плюсы:**
- Простота использования: хорошо задокументирована, много примеров, активное сообщество.
- Полная поддержка Telegram Bot API: поддерживает все функции API, включая inline-режим, вебхуки и многое другое.
- Асинхронность: полная поддержка asyncio, что позволяет эффективно обрабатывать множество запросов одновременно.
- Совместимость: легко интегрируется с асинхронными фреймворками, например, FastAPI.

**Минусы:**
- Переход на асинхронность: если раньше писал синхронно, придётся разобраться с async/await.

---

#### 2. `aiogram`

**Описание:**  
`aiogram` — современная асинхронная библиотека для создания Telegram-ботов, изначально проектировалась для asyncio. Очень производительная и гибкая.

**Плюсы:**
- Полная асинхронность: вся логика построена на asyncio.
- Высокая производительность: идеально подходит для ботов с высокой нагрузкой.
- Гибкость: легко создавать сложные кастомные решения.

**Минусы:**
- Крутая кривая обучения: нужно разбираться в асинхронном программировании.
- Документация менее обширная, чем у `python-telegram-bot`.

---

#### 3. `pyTelegramBotAPI` (TeleBot)

**Описание:**  
`pyTelegramBotAPI` — синхронная библиотека для создания Telegram-ботов. Простая и легкая в освоении.

**Плюсы:**
- Простота: отлично подходит для новичков и небольших проектов.
- Быстрый старт: можно сразу писать код, не заморачиваясь с async.

**Минусы:**
- Синхронность: не подходит для проектов с высокой нагрузкой или если нужно работать с асинхронным кодом.
- Масштабируемость ограничена.

---

#### 4. `Telethon`

**Описание:**  
`Telethon` — асинхронная библиотека, которая работает с Telegram API на уровне клиентов (userbot), а не только ботов. Позволяет делать сложные вещи.

**Плюсы:**
- Доступ к возможностям, которых нет в Bot API.
- Асинхронность на 100%.

**Минусы:**
- Сложность: требует хорошего понимания устройства Telegram.
- Использование userbots может противоречить правилам Telegram.


## **4. Создание простого эхо-бота**

image.png

```python
!pip install -q python-telegram-bot --upgrade  # Устанавливаем последнюю версию библиотеки для Telegram-бота

import nest_asyncio                         # Позволяет запускать asyncio-цикл в среде Colab
import asyncio                              # Стандартная библиотека Python для асинхронного кода

# Импортируем необходимые классы из python-telegram-bot
from telegram import Update                 # Представляет обновление (сообщение, команда и т.п.)
from telegram.ext import (
    ApplicationBuilder,                    # Строитель приложения — инициализирует бота
    CommandHandler,                        # Обрабатывает команды (например, /start)
    MessageHandler,                        # Обрабатывает обычные сообщения (текст и т.п.)
    ContextTypes,                          # Контекст, содержащий инфу о сообщении и пользователе
    filters,                               # Фильтры для выбора типа сообщений
)

from google.colab import userdata           # Позволяет получить секреты из user_data

nest_asyncio.apply()                        # Патчит loop, чтобы он мог работать в Colab

# Получаем токен бота из секрета (нужно заранее добавить TG_TOKEN в user_data)
BOT_TOKEN = userdata.get('TG_TOKEN')
if not BOT_TOKEN:
    raise ValueError("❌ Токен не найден! Добавь TG_TOKEN в Secrets через user_data.")

# Обработка команды /start
async def start(update: Update, context: ContextTypes.DEFAULT_TYPE):
    user = update.effective_user.first_name or "Пользователь"  # Получаем имя пользователя
    msg = "Привет! Я эхо-бот. Напиши что-нибудь — и я повторю!"  # Текст приветствия
    await update.message.reply_text(msg)                         # Отправляем ответ в Telegram
    print(f"{user} написал: /start")                             # Лог в Colab
    print(f"Бот ответил: {msg}\n")                               # Лог в Colab

# Обработка обычных текстовых сообщений
async def echo(update: Update, context: ContextTypes.DEFAULT_TYPE):
    user = update.effective_user.first_name or "Пользователь"   # Получаем имя пользователя
    msg = update.message.text                                   # Получаем текст сообщения
    await update.message.reply_text(msg)                        # Повторяем сообщение пользователю
    print(f"{user} написал: {msg}")                             # Лог в Colab
    print(f"Бот ответил: {msg}\n")                              # Лог в Colab

# Основная функция запуска бота
async def main():
    app = ApplicationBuilder().token(BOT_TOKEN).build()         # Создаём приложение Telegram-бота
    app.add_handler(CommandHandler("start", start))             # Обработчик команды /start
    app.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, echo))  # Обработчик текстовых сообщений (кроме команд)
    await app.run_polling()                                     # Запускаем бота в режиме опроса
```

```python
await main()                                                    # Запускаем всё приложение
```

## **5. Создание простого ChatGPT бота**

image.png

```python
# Устанавливаем зависимости: библиотеку для Telegram-бота, OpenAI SDK и патч для asyncio
!pip install -q python-telegram-bot --upgrade openai==1.74.0 nest_asyncio

import nest_asyncio                             # Патч для запуска asyncio-цикла в Colab (Jupyter)
import asyncio                                  # Стандартная библиотека Python для асинхронного кода
from telegram import Update                     # Класс Update, содержащий информацию о сообщении/команде
from telegram.ext import (
    ApplicationBuilder,                         # Помогает инициализировать Telegram-бота (асинхронно)
    CommandHandler,                             # Обработчик команд (/start, /help и т.д.)
    MessageHandler,                             # Обработчик обычных текстовых сообщений
    ContextTypes,                               # Объект с контекстом (содержит данные о сессии и т.п.)
    filters,                                    # Набор фильтров для MessageHandler (например, только текст)
)
from openai import OpenAI                        # Клиент OpenAI для запросов к ChatGPT
from google.colab import userdata                # Доступ к секретам, сохранённым в Colab (user_data)

# Применяем патч: позволяет вызывать asyncio внутри Colab без ошибок "Loop already running"
nest_asyncio.apply()

TG_TOKEN = userdata.get("TG_TOKEN")               # Токен Telegram-бота, сохранённый в Secrets → user_data["TG_TOKEN"]
OPENAI_API_KEY = userdata.get("OPENAI_API_KEY")   # Ключ OpenAI, сохранённый в Secrets → user_data["OPENAI_API_KEY"]

if not TG_TOKEN:
    raise ValueError("❌ TG_TOKEN не найден! Добавь его через Secrets (user_data).")
if not OPENAI_API_KEY:
    raise ValueError("❌ OPENAI_API_KEY не найден! Добавь его через Secrets (user_data).")

# 📡 Инициализируем клиента OpenAI с загруженным API-ключом
client = OpenAI(api_key=OPENAI_API_KEY)

# 📩 Обработчик команды /start: приветствует пользователя и объясняет, что бот делает
async def start(update: Update, context: ContextTypes.DEFAULT_TYPE):
    text = (
        "Привет! Я психолог Василий Степанович, работающий на основе ChatGPT.\n"
        "Давай обсудим твою проблему.."
    )
    await update.message.reply_text(text)                                           # Отправляем текстовое сообщение в Telegram

# 📩 Обработчик команды /payment: приветствует пользователя и объясняет, что бот делает
async def payment(update: Update, context: ContextTypes.DEFAULT_TYPE):
    text = (
        "Извините, раздел оплаты в разработке.\n"
    )
    await update.message.reply_text(text)

# 📩 Обработчик любых текстовых сообщений с сохранением истории
async def chatgpt_reply(update: Update, context: ContextTypes.DEFAULT_TYPE):
    user_text = update.message.text
    user = update.effective_user.first_name or "Пользователь"
    print(f"{user} написал: {user_text}")

    # Если впервые, создаём пустой список для истории
    if "history" not in context.user_data:
        context.user_data["history"] = [
            {"role": "system", "content": "Ты — опытный психолог Василий Степанович, узнай проблему и помоги пользователю решить ее."}
        ]

    # Добавляем реплику пользователя в историю
    context.user_data["history"].append({"role": "user", "content": user_text})

    try:
        # Отправляем в OpenAI полную историю
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=context.user_data["history"],  # ← здесь уже полный контекст
            temperature=0.7,
            max_tokens=512,
        )
        reply = response.choices[0].message.content.strip()

        # Добавляем ответ бота в историю
        context.user_data["history"].append({"role": "assistant", "content": reply})

    except Exception as e:
        reply = "Извини, произошла ошибка при запросе к OpenAI."
        print("Ошибка OpenAI:", e)

    await update.message.reply_text(reply)
    print(f"Бот ответил: {reply}\n")

# 🎯 Основная функция запуска бота: создаём приложение, регистрируем обработчики и запускаем polling
async def main():
    app = ApplicationBuilder().token(TG_TOKEN).build()                              # Инициализируем Telegram-приложение с токеном
    app.add_handler(CommandHandler("start", start))                                 # Регистрируем обработчик команды /start
    app.add_handler(CommandHandler("payment", payment))                             # Регистрируем обработчик команды /payment
    app.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, chatgpt_reply)) # Регистрируем обработчик всех текстовых сообщений (исключая команды): отвечает через ChatGPT
    await app.run_polling()                                                         # Запускаем бота в режиме опроса (polling)
```

```python
# 🚀 Запускаем
await main()
```
